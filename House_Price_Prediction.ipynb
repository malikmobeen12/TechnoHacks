```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV files into DataFrames
test_df = pd.read_csv(r'C:\Users\mobee\OneDrive\Desktop\Internship\Predicting House Price\test.csv')
train_df = pd.read_csv(r'C:\Users\mobee\OneDrive\Desktop\Internship\Predicting House Price\train.csv')

# Display the first 5 rows
print("First 5 rows of training data:")
print(train_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nFirst 5 rows of test data:")
print(test_df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print("\nTraining data column names and types:")
print(train_df.info())

print("\nTest data column names and types:")
print(test_df.info())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Drop columns with more than 20% missing values
null_threshold = 0.20
train_df = train_df.loc[:, train_df.isnull().mean() < null_threshold]
test_df = test_df.loc[:, test_df.isnull().mean() < null_threshold]

# Drop the 'Id' column
train_df.drop(columns=['Id'], inplace=True)
test_ids = test_df['Id']  # Store for submission
test_df.drop(columns=['Id'], inplace=True)

# Separate target variable from training features
y_train = train_df.pop('SalePrice')

# Select only numeric columns for training and testing
X_train = train_df.select_dtypes(include=['int64', 'float64'])
X_test = test_df.select_dtypes(include=['int64', 'float64'])

# Train a Random Forest Regressor model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Generate predictions on the test set
y_pred = model.predict(X_test)

# Create a submission DataFrame
submission_df = pd.DataFrame({'Id': test_ids, 'SalePrice': y_pred})

# Write the submission to a CSV file
submission_df.to_csv('submission.csv', index=False)
```

```python
# Impute missing values with the mean of each column
for col in X_train.columns:
    X_train[col].fillna(X_train[col].mean(), inplace=True)
    X_test[col].fillna(X_train[col].mean(), inplace=True)  # Use train mean for test

# Train a Random Forest Regressor model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Generate predictions on the test set
y_pred = model.predict(X_test)

# Create a submission DataFrame
submission_df = pd.DataFrame({'Id': test_ids, 'SalePrice': y_pred})

# Write the submission to a CSV file
submission_df.to_csv('submission.csv', index=False)
```
